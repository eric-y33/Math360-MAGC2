# Problem Statement

Construct a mathematical model to estimate the **total annual income** of Loafe Café at both locations Alumni Centre and Sauder as a function of daily temperature, daily precipitation, daily humidity and frequency of customers.

# Variables and Parameters

| Symbol | Description | Type | Dimension | Units |
|---|---|---|---|---|
| $d$ | day index within a (non-leap) year, $d \in \{1, \dots, 365\}$ | independent variable | 1 | — |
| $h$ | hour index (start of hour) | independent variable | 1 | hour-of-day |
| $l$ | location (Alumni, Sauder) | independent variable (categorical) | 1 | — |
| $T$ | daily average temperature on day $d$ | independent variable (weather input) | $T$ | °C |
| $P$ | daily total precipitation on day $d$ | independent variable (weather input) | $L$ | mm |
| $H$ | daily average humidity on day $d$ | independent variable (weather input) | 1 | % |
| $C$ | expected total customers on day $d$ (both locations) | dependent variable (model output) | 1 | customers/day |
| $s$ | average spend per customer (assumed constant or scenario value) | parameter | 1 | CAD/customer |
| $R$ | total annual income | derived aggregate output | 1 | CAD/year |
| $T_0$ | reference temperature level | parameter (reference level) | $T$ | °C |
| $P_0$ | reference precipitation level | parameter (reference level) | $L$ | mm |
| $H_0$ | reference humidity level | parameter (reference level) | 1 | % |
| $C_0$ | baseline customers at reference weather $(T_0, P_0, H_0)$ | parameter | 1 | customers/day |
| $a$ | rate of change of customers w.r.t temperature (squared) | parameter (sensitivity) | $T^{-2}$ | customers / (°C)$^2$ |
| $b$ | rate of change of customers w.r.t precipitation | parameter (sensitivity) | $L^{-1}$ | customers / mm |
| $c$ | rate of change of customers w.r.t humidity (squared) | parameter (sensitivity) | 1 | customers / (%)$^2$ |


# Assumptions and Constraints:


- The number of customers is non-negative and proportional to temperature squared, precipitation, and humidity squared. This means that the number of customers change in the same direction as temperature and humidity deviate further from a reference value, whereas the number changes in either the same or opposite direction as precipitation increases or decreases.
- Annual revenue linearly scales with the number of customers per year.
- The amount of money individual customers spend remains constant all year round.
- On average, customers at the Alumni Centre spend the same amount of money as customers at the Sauder building. This allows us to combine the customer data of the two.
- To evaluate the potential correlation of weather to customer visits, we do not account for external factors such as the school year schedule.
- For usage of this model in current and future years, we assume that weather trends each month remain the same in 2025 and onwards.
- We assume weather trends are different each month, so we will construct separate functions for each to estimate customer visits more accurately given new weather data during a particular month.


# Building the Solution:

We estimate total annual income from weather by:
1) loading the provided datasets and doing some light explanatory data analysis,  
2) learning monthly weather distributions with KDEs,  
3) specifying a transparent customer response $ C(T,P,H) $ and estimating its parameters from January/July,  
4) simulating a typical year (weather draws month-by-month) to get revenue, and  
5) propagating uncertainty via Monte Carlo.

> We intentionally keep the model low-dimensional and interpretable because only **two** customer months (Jan/Jul) are available.

**Load data and quick EDA**

We read the customer and weather files using our original schema.  
To keep the notebook compact, we only display a few top rows.  
We also precompute **Jan/Jul totals across both locations per hour** (this is what later becomes daily totals per month).

```python
plt.rcParams["figure.figsize"] = (10, 3)
plt.rcParams["axes.grid"] = True

# load data
cust_data = pd.read_csv("customer_data_2025.csv")
cust = pd.DataFrame(cust_data)

weather_data = pd.read_csv("weatherstats_vancouver_daily_2025.csv")  
weather = pd.DataFrame(weather_data)

# limited peeks
print('---- Customer data (head with first 5 rows) ----')
display(cust.head(5))
print('---- Weather data (head with first 5 rows)----')
display(weather.head(5))

# jan/jul totals across both locations
JAN_AL = "January_Frequency_Customers_Alumni_Centre"
JAN_SD = "January_Frequency_Customers_Sauder"
JUL_AL = "July_Frequency_Customers_Alumni_Centre"
JUL_SD = "July_Frequency_Customers_Sauder"

cust["Jan_total"] = cust[JAN_AL].fillna(0) + cust[JAN_SD].fillna(0)
cust["Jul_total"] = cust[JUL_AL].fillna(0) + cust[JUL_SD].fillna(0)
```

**Build monthly KDEs and visualize Jan vs Jul**

We split **weather** by month and fit simple Gaussian KDEs for temperature, precipitation, and humidity.  
KDEs give smooth seasonal distributions without committing to a specific parametric family.

**Figure 1** overlays **January vs July** PDFs for \(T,P,H\) so we can visually confirm winter/summer regimes that will drive the customer model.


```python
# ========= Monthly splits & KDEs =========
# split weather into 12 months
listOfMonthData = []
for i in range(1, 13):
    listOfMonthData.append(weather.loc[weather["Month"] == i, ["temperature", "precipitation", "humidity"]].dropna())

class Month:
    def __init__(self, name, month_pdf_T, month_pdf_P, month_pdf_H):
        self.name = name
        self.month_pdf_T = month_pdf_T
        self.month_pdf_P = month_pdf_P
        self.month_pdf_H = month_pdf_H

# build KDEs per month
listOfMonthPDFs = []
for i in range(12):
    month = listOfMonthData[i]
    pdf_T = stats.gaussian_kde(month["temperature"])
    pdf_P = stats.gaussian_kde(month["precipitation"])
    pdf_H = stats.gaussian_kde(month["humidity"])
    listOfMonthPDFs.append(Month(str(i+1), pdf_T, pdf_P, pdf_H))

def _plot_overlay(ax, kde_a, kde_b, grid, label_a, label_b, title, xlabel):
    ax.plot(grid, kde_a(grid), label=label_a)
    ax.plot(grid, kde_b(grid), label=label_b)
    ax.set_title(title); ax.set_xlabel(xlabel); ax.legend()

# Figure 1 (overlays Jan vs Jul for T,P,H)
fig, axes = plt.subplots(1, 3, figsize=(12, 3))
gT = np.linspace(-15, 30, 400)
gP = np.linspace(0, 60, 400)
gH = np.linspace(20, 100, 400)
_plot_overlay(axes[0], listOfMonthPDFs[0].month_pdf_T, listOfMonthPDFs[6].month_pdf_T, gT, "Jan", "Jul", "Temperature PDFs", "°C")
_plot_overlay(axes[1], listOfMonthPDFs[0].month_pdf_P, listOfMonthPDFs[6].month_pdf_P, gP, "Jan", "Jul", "Precipitation PDFs", "mm")
_plot_overlay(axes[2], listOfMonthPDFs[0].month_pdf_H, listOfMonthPDFs[6].month_pdf_H, gH, "Jan", "Jul", "Humidity PDFs", "%")
plt.suptitle("Figure 1 — Monthly KDE overlays (Jan vs Jul)", y=1.1)
plt.tight_layout()
```

**Specify $C(T,P,H)$ and estimate $a,b,c,C_0$ from Jan/Jul**


$$C(T,P,H)=\max\{\,C_0 + a\,(T-T_0)^2 - b\,(P-P_0) - c\,(H-H_0)^2,\ 0\,\}$$

- $T_0,P_0,H_0$ are long-run averages from the weather data.  
- We get **Jan/Jul mean weather** from the KDEs (numerical mean via integration on a fine grid).  
- We estimate $a,b,c$ using your finite-difference idea based on the **change in average daily customers** from January to July.  
- $C_0$ is set to the midpoint of Jan/Jul daily averages (a stable baseline given only two labeled months).


```python
# ========= Customer model c(T,P,H) and parameter estimates =========
# reference levels (long-run means)
t0 = np.mean(weather["temperature"])
p0 = np.mean(weather["precipitation"])
h0 = np.mean(weather["humidity"])

# function to extract mean from a KDE on a grid
def kde_mean(kde, grid):
    pdf_vals = kde(grid)
    pdf_vals = pdf_vals / np.trapz(pdf_vals, grid)  # normalize just in case
    return np.trapz(grid * pdf_vals, grid)

# Jan/Jul means from KDEs
jan_mean_temp = kde_mean(listOfMonthPDFs[0].month_pdf_T, gT)
jul_mean_temp = kde_mean(listOfMonthPDFs[6].month_pdf_T, gT)
jan_mean_precip = kde_mean(listOfMonthPDFs[0].month_pdf_P, gP)
jul_mean_precip = kde_mean(listOfMonthPDFs[6].month_pdf_P, gP)
jan_mean_humid = kde_mean(listOfMonthPDFs[0].month_pdf_H, gH)
jul_mean_humid = kde_mean(listOfMonthPDFs[6].month_pdf_H, gH)

# daily customers in Jan/Jul from hourly totals
JAN_DAYS = 31
JUL_DAYS = 31
January_Total_Customers_Loafe = cust["Jan_total"].sum()
July_Total_Customers_Loafe = cust["Jul_total"].sum()
Avg_Daily_Customers_Jan = January_Total_Customers_Loafe / JAN_DAYS
Avg_Daily_Customers_Jul = July_Total_Customers_Loafe / JUL_DAYS
avg_daily_customer_change = Avg_Daily_Customers_Jul - Avg_Daily_Customers_Jan

# estimate a, b, c 
a = avg_daily_customer_change / ((jul_mean_temp - t0)**2 - (jan_mean_temp - t0)**2)
b = avg_daily_customer_change / ((jul_mean_precip - p0) - (jan_mean_precip - p0))
c = avg_daily_customer_change / ((jul_mean_humid - h0)**2 - (jan_mean_humid - h0)**2)

# baseline guess
c0 = 0.5*(Avg_Daily_Customers_Jan + Avg_Daily_Customers_Jul)

print(f"Ref means: T0={t0:.2f}°C, P0={p0:.2f}mm, H0={h0:.2f}%")
print(f"Jan vs Jul daily customers: {Avg_Daily_Customers_Jan:.1f} → {Avg_Daily_Customers_Jul:.1f}")
print(f"a={a:.3f}, b={b:.3f}, c={c:.3f}, C0={c0:.1f}")

# model
def cFunc(TempDay, PrecipitationDay, HumidityDay):
    val = c0 + a*(TempDay - t0)**2 - b*(PrecipitationDay - p0) - c*(HumidityDay - h0)**2
    return max(val, 0.0)
```

## Yearly Simulation

We simulate a **typical** (non-leap) year by:
- mapping day $1..365$ to its **month**,  
- drawing a single $(T,P,H)$ sample from that month’s KDEs,  
- applying $C(T,P,H)$, and  
- multiplying by a constant **spend per customer** ($15; adjustable).

This produces a **point estimate** of annual revenue.

```python
days_per_month = np.array([31,29,31,30,31,30,31,31,30,31,30,31])  
month_edges = np.cumsum(days_per_month)

def dayToMonth(day):
    return int(np.searchsorted(month_edges, day))  # 0..11

def dailyRev(month, spend_per_customer=8.0):
    T_hat = listOfMonthPDFs[month].month_pdf_T.resample(1).ravel()[0]
    P_hat = listOfMonthPDFs[month].month_pdf_P.resample(1).ravel()[0]
    H_hat = listOfMonthPDFs[month].month_pdf_H.resample(1).ravel()[0]
    customersDay = cFunc(T_hat, P_hat, H_hat)
    return customersDay * spend_per_customer

def yearlyRev(spend_per_customer=15.0):
    total = 0.0
    for i in range(1, 365+1):
        m = dayToMonth(i)  # 0..11
        total += dailyRev(m, spend_per_customer=spend_per_customer)
    return total

yearRev = yearlyRev(spend_per_customer=15.0)
print(f"Point estimate — revenue/year (s=$8): ${yearRev:,.0f}")
```

## Monte Carlo Estimation

To acknowledge limited Jan/Jul labels, we run a **Monte Carlo**:
- perturb $a,b,c,C_0,T_0,P_0,H_0$ by adding modest Gaussian noise,  
- resample weather each day from the month’s KDEs,  
- compute the annual revenue, and repeat $N=1000$ times.

We then visualize the distribution (Figure 3) and report the median and a 90% interval.

```python
# parameter sigmas 
a_err, b_err, c_err = 0.5, 2.0, 0.2
t0_err, p0_err, h0_err, c0_err = 1.0, 2.0, 3.0, 30.0

def dailyRevMC(month, rng, spend_per_customer=8.0):
    a_s = a + rng.normal(0.0, a_err)
    b_s = b + rng.normal(0.0, b_err)
    c_s = c + rng.normal(0.0, c_err)
    t0_s = t0 + rng.normal(0.0, t0_err)
    p0_s = p0 + rng.normal(0.0, p0_err)
    h0_s = h0 + rng.normal(0.0, h0_err)
    c0_s = c0 + rng.normal(0.0, c0_err)

    T_hat = listOfMonthPDFs[month].month_pdf_T.resample(1).ravel()[0]
    P_hat = listOfMonthPDFs[month].month_pdf_P.resample(1).ravel()[0]
    H_hat = listOfMonthPDFs[month].month_pdf_H.resample(1).ravel()[0]

    val = c0_s + a_s*(T_hat - t0_s)**2 - b_s*(P_hat - p0_s) - c_s*(H_hat - h0_s)**2
    customersDay = max(val, 0.0)
    return customersDay * spend_per_customer

def yearlyRevMC(N=1000, spend_per_customer=8.0, seed=7):
    rng = np.random.default_rng(seed)
    total = 0.0
    for i in range(1, 365+1):
        m = dayToMonth(i)
        total += dailyRevMC(m, rng, spend_per_customer=spend_per_customer)
    return total

N = 1000
yearlyRevenues = [yearlyRevMC() for _ in range(N)]
```

- **Figure 3:** Histogram of annual revenue from Monte Carlo.  
- **Figure 4:** Simple one-dimensional sensitivities of expected customers vs each weather variable around the reference levels (helps interpret $a,b,c$).  
We also print the median and 90% interval of revenue.

```python
# ========= Plot MC histogram & sensitivity mini-panel =========
fig, ax = plt.subplots(1, 1, figsize=(10, 3))
ax.hist(yearlyRevenues, density=True, bins=30)
ax.set_title("Figure 3 — Annual revenue uncertainty (Monte Carlo)")
ax.set_xlabel("Revenue (CAD)"); ax.set_ylabel("Density")
plt.tight_layout()

# Figure 4 — simple 1D sensitivities (T,P,H) around reference (overlays in one figure)
fig, axes = plt.subplots(1, 3, figsize=(12, 3))
def _sens_1d(ax, var):
    if var == "T":
        xs = np.linspace(t0-15, t0+15, 200)
        ys = [max(c0 + a*(x - t0)**2 - b*(p0 - p0) - c*(h0 - h0)**2, 0) for x in xs]
        ax.plot(xs, ys); ax.set_xlabel("Temperature (°C)")
    elif var == "P":
        xs = np.linspace(0, max(60, p0+30), 200)
        ys = [max(c0 + a*(t0 - t0)**2 - b*(x - p0) - c*(h0 - h0)**2, 0) for x in xs]
        ax.plot(xs, ys); ax.set_xlabel("Precipitation (mm)")
    else:
        xs = np.linspace(20, 100, 200)
        ys = [max(c0 + a*(t0 - t0)**2 - b*(p0 - p0) - c*(x - h0)**2, 0) for x in xs]
        ax.plot(xs, ys); ax.set_xlabel("Humidity (%)")
    ax.set_ylabel("Expected customers/day")

_sens_1d(axes[0], "T")
_sens_1d(axes[1], "P")
_sens_1d(axes[2], "H")
plt.suptitle("Figure 4 — Local sensitivities around reference weather", y=1.1)
plt.tight_layout()

# summary stats
yr = np.array(yearlyRevenues)
q5, q50, q95 = np.percentile(yr, [5, 50, 95])
print(f"MC median: ${q50:,.0f} | 90% CI: [${q5:,.0f}, ${q95:,.0f}]")
```
